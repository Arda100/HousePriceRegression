## 2. Exploratory Data Analysis (EDA)

```python
sns.histplot(x="SalePrice", data=data, kde=True)
plt.title("Histogram of Sale Prices")
plt.xlabel("Sale Price")
plt.ylabel("Frequency")
plt.show()
```

```python
saleprices_log = np.log(data['SalePrice'])
sns.histplot(x=saleprices_log, kde=True)
plt.title("Histogram of Sale Prices' Logharitms")
plt.xlabel("Logharitm of Sale Price")
plt.ylabel("Frequency")
plt.show()
```

```python
top_correlation_features = data.corr(numeric_only=True)['SalePrice'].drop('SalePrice').abs().nlargest(15).index.tolist() + ['SalePrice']

plt.figure(figsize=(12, 10))
sns.heatmap(data[top_correlation_features].corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix for Most Correlated 15 Feature With Sale Price")
plt.show()
```

```python
sns.boxplot(x='OverallQual', y='SalePrice', data=data)
plt.title("Boxplot for OverallQual")
plt.show()
```

## 3. Data Cleaning

```python
null_counts = data.isnull().sum()
columns_with_nulls = null_counts[null_counts != 0]
print(columns_with_nulls)
```

```python
data.drop(['PoolQC', 'MiscFeature', 'Alley', 'Fence'], axis=1, inplace=True)
# These features were deleted because they have low variance
```

```python
columns = ['LotFrontage', 'MasVnrType', 'MasVnrArea', 'BsmtQual', 'BsmtCond',
           'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2', 'Electrical', 'FireplaceQu',
           'GarageType', 'GarageYrBlt', 'GarageFinish', 'GarageQual', 'GarageCond']

print("--- Unique Values ---")
print("-" * 50)

for col in columns:
    print(f"Column: '{col}'")
    
    print(data[col].unique())
    
    print("-" * 50)
```

```python
def fill_empty(column, value):
    data[column] = data[column].fillna(value)
```

```python
garage_cols = ['GarageType', 'GarageFinish', 'GarageQual', 'GarageCond']
for col in garage_cols:
    fill_empty(col, 'None')
```

```python
garage_numeric_cols = ['GarageYrBlt', 'GarageArea', 'GarageCars']
for col in garage_numeric_cols:
    fill_empty(col, 0)
```

```python
basement_cols = ['BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2']
for col in basement_cols:
    fill_empty(col, 'None')
```

```python
fill_empty('FireplaceQu', 'None')
fill_empty('MasVnrType', data['MasVnrType'].mode()[0])
fill_empty('MasVnrArea', 0)
fill_empty('Electrical', data['Electrical'].mode()[0])
data['LotFrontage'] = data.groupby('Neighborhood')['LotFrontage'].transform(
    lambda x: x.fillna(x.median())
)
```

```python
null_counts = data.isnull().sum()
columns_with_nulls = null_counts[null_counts != 0]
print(columns_with_nulls)
```
